```python
#drop uniportant variables 
drop_cols = ['id_produto', 'id_contrato', 'id_cliente', 'data_entrada','data_entrada.1', 'data_final'] 
df_clean = df.drop(drop_cols, axis=1)
```

```python
#criando variáveis quantitativas
# razão valor financiado/faturamento
df['fin_fat'] = df['valor_fin']/df['fat']

# log do faturamento
df['fat_log'] = np.log(df['fat'])

# prazo 
## df['prazo_dias'] = df['data_entrada'] - df['data_final'] 
```

```python
#separando as variáveis por tipo (quantitativa e categórica)
#variáveis categóricas
quant_var = ['valor_fin','qtde_prest', 'juros_efet', 'provisao','fat', 'valor_garantia','fin_fat', 'fat_log']
df_cat_var = df_clean.drop(quant_var, axis=1)
```

```python
# variáveis quantitativas
df_quant_var = df_clean.drop(df_cat_var, axis=1)
```
