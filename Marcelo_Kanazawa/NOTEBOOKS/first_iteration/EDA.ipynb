```python
#import libraries
import os 
import pandas as pd # pacote para manipular os dados
import numpy as np # pacote para numeros 
import matplotlib.pyplot as plt # pacote para visualização dos dados
import seaborn as sns  # pacote para visualização dos dados
```

### Read Data

```python
df = pd.read_csv('../DATABASE/base_estudos.csv') # Lendo a base de dados csv
```

```python
colunas = list(df.columns) # Lendo as colunas do dataframe original e passando para o objeto tipo lista chamado colunas
print(colunas)
```

```python
#criando variáveis quantitativas
# razão valor financiado/faturamento
df['razao_fin.fat'] = df['valor_fin']/df['fat']

# log do faturamento
df['log_fat'] = np.log(df['fat'])

# prazo 
df['prazo_dias'] = df['data_final'] - df['data_entrada']
```

```python
#drop unimportant variables 
drop_cols = ['id_produto', 'id_contrato', 'id_cliente', 'data_entrada','data_entrada.1', 'data_final'] 
df_clean = df.drop(drop_cols, axis=1) # Função para deletar uma coluna do dataframe escolhido
```

# Create a report --- PDF
---

```python
from jinja2 import Environment, FileSystemLoader
env = Environment(loader=FileSystemLoader('.'))
template = env.get_template("my_report.html")
```

```python
 pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df_describe = df_clean.describe()
```

```python
df_describe_1 = df_describe.iloc[:,:4].round(3)
df_describe_2 = df_describe.iloc[:,4:].round(3)
```

```python
display(df_describe_1)
```

```python
display(df_describe_2)
```

```python
df_corr = df.corr()
```

```python
template_vars = {"title" : "Data Frame stats",
                 "describe_1": df_describe_1.to_html(),
                 "describe_2": df_describe_2.to_html(),
                 "df_corr": df_corr.to_html()}
```

```python
html_out = template.render(template_vars)
```
